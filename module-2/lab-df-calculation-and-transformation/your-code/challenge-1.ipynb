### Print the distinct values in `Type 1` and `Type 2` combined

```python
# enter your code here
distinct_val=pokemon.groupby(['Type 1','Type 2']).size().reset_index().rename(columns={0:'distinct count'})
print('Distinct values: ',len(distinct_val))
display(distinct_val.head(10))
```

Check out the pokemon names in the first 10 rows. You find there are junk texts in the pokemon names which contain "Mega". For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

### Cleanup `Name` that contain "Mega"

```python
# enter your code here
pokemon['Name']=pokemon['Name'].str.replace('Mega','')

# test transformed data
pokemon.head(10)
```

### Create a new column called `A/D Ratio` whose value equals to `Attack` devided by `Defense`

For instance, pokemon #1 has the Attack score 49 and Defense score 49. The corresponding `A/D Ratio` is 49/49=1.

```python
# enter your code here
pokemon['A/D Ratio']=pokemon['Attack']/pokemon['Defense']

# test transformed data

pokemon.head(10)
```

### Print the pokemon with the highest `A/D Ratio`

```python
# enter your code here
pokemon.loc[pokemon['A/D Ratio']==pokemon['A/D Ratio'].max()][['Name','A/D Ratio']]
```

### Print the pokemon with the lowest A/D Ratio

```python
# enter your code here
pokemon.loc[pokemon['A/D Ratio']==pokemon['A/D Ratio'].min()][['Name','A/D Ratio']]
```

### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Conditions:

* If `Type 2` value is a valid string, the `Combo Type` value should be `<Type 1>-<Type 2>` (e.g. `Grass-Poison`).

* If `Type 2` value is `NaN`, the `Combo Type` value should be the same as `Type 1` which always exists.

*Hint: Consider using function and `apply`.*

```python
# enter your code here

pokemon['Combo Type']=np.where(pokemon['Type 2'].isnull(),pokemon['Type 1'],pokemon['Type 1']+"-"+pokemon['Type 2'])

# test transformed data

pokemon.head(10)
```

```python
### Print `Combo Type` for pokemons whose `A/D Ratio` are among the top 5 
```

```python
# enter your code here
pokemon.nlargest(5,'A/D Ratio')[['Combo Type','A/D Ratio']]
```

### For the 5 `Combo Type` values printed from the previous question, calculate the aggregated `Attack` scores for each `Combo Type`.

```python
# enter your code here
top5combo=pokemon.nlargest(5,'A/D Ratio')['Combo Type'].tolist()
avg_attk=pokemon.groupby('Combo Type').agg({'Attack':'mean'})
display(avg_attk.loc[top5combo])
```

Pokemon Total attribute is the sum of attributes **HP, Attack, Defense, Sp. Atk, Sp. Def and Speed** > **Total**

```python
# enter your code here
formula=pokemon['Attack']+pokemon['Defense']+pokemon['Sp. Atk']+pokemon['Sp. Def']+pokemon['Speed']+pokemon['HP']
```

To check my hipothesis I'm going to obtain the difference between my series **formula** and attribute **Total**<br>

```python
print("Sum of Total and formula differences: ",sum(pokemon['Total']-formula))
if sum(pokemon['Total']-formula)==0:
    print("Hipothesis confirmed "+u'\u2713')
```

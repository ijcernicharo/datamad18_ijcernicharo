### Question 1: How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?

```python
# Driver 1 enter codes here
stepsByMonth=fitbit.pivot_table(index="Months_encoded", aggfunc=({'Steps': ("mean", "median")}))
stepsByMonth['difference']= ["0"]+[str(j-i) for i,j in zip(stepsByMonth.iloc[:-1, 0],stepsByMonth.iloc[1:, 0])]
stepsByMonth['difference']=stepsByMonth['difference'].astype("float64")
stepsByMonth.columns=stepsByMonth.columns.droplevel()
stepsByMonth.columns=["Mean", "Median", "Difference"]
fig, ax = plt.subplots(2,1, figsize=(15,10))
stepsByMonth.plot(y='Difference', ax=ax[0], marker="o", clip_on=False)
stepsByMonth.plot(y='Median', ax=ax[1],color="r",marker="o", clip_on=False)
ax[0].grid()
ax[1].grid()
ax[0].set_title("Average number of steps change by month")
ax[1].set_title("Median steps by month")
ax[0].set_xticks(range(1,13))
ax[0].set_xlim(1,12)
ax[1].set_xticks(range(1,13))
ax[1].set_xlim(1,12)
```

### Question 2: Is Fitbitter more active on weekend or workdays?

```python
activityPerWeekDay=fitbit.pivot_table(index="Work_or_Weekend", aggfunc=({"Steps":"mean", "Minutes Lightly Active":"mean", 'Minutes Fairly Active':"mean",'Minutes Very Active':"mean"}))
activityPerWeekDay["Total minutes active"]=activityPerWeekDay["Minutes Lightly Active"]+activityPerWeekDay['Minutes Fairly Active']+activityPerWeekDay['Minutes Very Active']
```

```python
activityPerWeekDay1=fitbit.pivot_table(index="Days_encoded", aggfunc=({"Steps":"mean", "Minutes Lightly Active":"mean", 'Minutes Fairly Active':"mean",'Minutes Very Active':"mean"}))
activityPerWeekDay1["Total minutes active"]=activityPerWeekDay1["Minutes Lightly Active"]+activityPerWeekDay1['Minutes Fairly Active']+activityPerWeekDay1['Minutes Very Active']
per_day=activityPerWeekDay1[["Total minutes active", "Steps"]]
fig= plt.figure(figsize=(15,5))
ax=plt.gca()
per_day['Total minutes active'].plot(ax=ax,color='red',marker="o", clip_on=False)
ax1=ax.twinx()
per_day['Steps'].plot(ax=ax1,color='blue',marker="o", clip_on=False)
ax.legend(loc='upper left')
ax.set_xticks(range(0,7))
ax.set_xticklabels(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
ax.grid()
ax.set_ylabel('Total Minutes Active',fontweight='bold')
ax1.set_ylabel('Steps',fontweight='bold')
ax1.legend(loc='upper right')
ax.set_xlim(0,6)
display(activityPerWeekDay[["Total minutes active", "Steps"]])
```

We have got to a point where there is a discussion between the concept of being more active by taking more steps or the amount of time you are active. We have decided to bet on the total minutes active on, reaching the conclusion that weekdays are more active than weekend.

# Challenge 1

#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

fig,axes=plt.subplots(2)

axes[0].plot(x,y)
axes[0].grid()

axes[1].plot(x,z)
axes[1].grid()
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig,(axes1,axes2)=plt.subplots(2)

axes1.plot(x,y)
axes1.grid()

axes2.plot(x,z)
axes2.grid()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig,(axes1,axes2)=plt.subplots(2,figsize=(10,7))

axes1.plot(x,y)
axes1.grid()

axes2.plot(x,z)
axes2.grid()
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig,(axes1,axes2)=plt.subplots(2,figsize=(10,7),sharex=True)

axes1.plot(x,y)
axes1.grid()
axes1.set_ylabel('$2·x$',fontsize=15)

axes2.plot(x,z)
axes2.grid()
axes2.set_ylabel(r'$x^2$',fontsize=15)
axes2.set_xlabel('Varible x',fontsize=15)
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig,(axes1,axes2,axes3)=plt.subplots(nrows=3,figsize=(15,12))

axes1.plot(x,y)
axes1.grid()
axes1.set_ylabel('$x^2$',fontsize=15)
axes1.set_xlabel('Variable x',fontsize=15)

axes2.plot(x,z)
axes2.grid()
axes2.set_ylabel(r'$e^x$',fontsize=15)
axes2.set_xlabel('Variable x',fontsize=15)
axes2.set_yscale('log')
axes2.set_xscale('log')

axes3.loglog(x,z)
axes3.grid()
axes3.set_ylabel(r'$e^x$',fontsize=15)
axes3.set_xlabel('Variable x',fontsize=15)
```

#### Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in this repo whose filenames contain .scavengerhunt. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching on your computer.

```python
import requests
import json
import pandas as pd
import os
from pandas.io.json import json_normalize
import base64
```

```python
api='https://api.github.com/repos/'
user='ironhack-datalabs/madrid-oct-2018/'
what='git/trees/master'
how='?recursive=1'
completerequest=api+user+what+how
response = requests.get(completerequest)
result = response.json()
```

___

# Two kinds of approx
    1. Easy path
    2. Hardcore path

___

## 1. Easy path

```python
scanvengers=[]
realpath=[]
scv=[]
for item in pd.DataFrame(result['tree'])['path']:
    if str(item).endswith('scavengerhunt'):
        scanvengers.append(item)
        realpath.append('/home/ivan/Desktop/IRONHACK_DATA_ANALYTICS/madrid-oct-2018/'+item)
        if '/' in item:
            scv.append(item.split('/')[-1])
        else:
            scv.append(item)
```

```python
final_df=pd.DataFrame(scanvengers,columns=['path'])
final_df['names']=scv
final_df['real path']=realpath
final_df=final_df[['names','path','real path']]
final_df=final_df.sort_values('names',ascending=True)
display(final_df)
```

### Last lines of code for the cold joke

```python
cold_joke=''
for pathy in final_df['real path']:
    varp=open(pathy,'r')
    while 1:
        line=varp.readline()
        if len(line)==0:
            break
        cold_joke+=' '+line.replace('\n','')
            
    varp.close()
```

```python
print("")
print(" >> Hidden Cold joke (Easter Egg): ")
print("")
print(chr(27)+"[1;34m"+cold_joke.upper()+chr(27)+"[0m"+"")
print("")
```

## 2. Hardcore path

```python
url=[]
scvurl=[]
resultdf=pd.DataFrame(result['tree'])
for i in range(len(resultdf)):
    if str(resultdf['path'][i]).endswith('scavengerhunt'):
        url.append(resultdf['url'][i])
        if '/' in resultdf['path'][i]:
            scvurl.append((resultdf['path'][i]).split('/')[-1])
        else:
            scvurl.append((resultdf['path'][i]))
```

```python
dataurl=pd.DataFrame(scvurl,columns=['name'])
dataurl['url']=url
dataurl=dataurl.sort_values('name',ascending=True).reset_index(drop=True)
display(dataurl)
```

```python
cold_joke_hard=''
for i in range(len(dataurl)):
    responseto=requests.get(dataurl['url'][i])
    if 'content' in responseto.json().keys():
        string=base64.b64decode(responseto.json()['content']).decode("utf-8").replace('\n','')
    else:
        print('You have exceeded the max number of requests, try again in one hour')
    print(i," >> ",string)
    cold_joke_hard+=' '+string 
```

```python
print("")
print(" >> Hidden Cold joke (Easter Egg) - HARD PARTY: ")
print("")
print(chr(27)+"[1;34m"+cold_joke_hard.upper()+chr(27)+"[0m"+"")
print("")
```

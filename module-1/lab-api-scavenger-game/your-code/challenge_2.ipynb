# Challenge 2

Count how many commits were made between October 20, 2018 and October 25, 2018.<br>

    1. Obtain all the commits made between October 20, 2018 and October 25, 2018 via API, which is a JSON array that contains multiple commit objects.

    2. Count how many commit objects are there in the array.

```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

___

1. Requesting data for the date interval between October 20 and October 25

```python
api='https://api.github.com/repos/'
user='ironhack-datalabs/madrid-oct-2018/'
what='commits'
since='?since=2018-10-20T00:00:00Z'
until='&until=2018-10-25T23:59:59Z'
limit='&per_page=100'
completerequest=api+user+what+since+until+limit
response = requests.get(completerequest)
result = response.json()
with open('challenge_2.json', 'w') as outfile:
    json.dump(result, outfile)
```

```python
commit=json_normalize(result)
display(commit.head(5))
```

___

 2. Count how many commit objects are there in the array.

```python
print("")
print(("Number of commits between October 20 and 25  >>>>  ").upper(),len(commit))
print("")
```

```python
Oct_20_25=commit[['commit.committer.date','commit.author.name','commit.message']].sort_values(['commit.committer.date'],ascending=True)
Oct_20_25.columns=['Date and Time','User name','Commit message']
display(Oct_20_25)
```

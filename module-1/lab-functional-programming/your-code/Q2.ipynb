```python
def get_bow_from_docs(docs, stop_words=[]):
    
    # In the function, first define the variables you will use such as `corpus`, `bag_of_words`, and `term_freq`.
    
    
    """
    Loop `docs` and read the content of each doc into a string in `corpus`.
    Remember to convert the doc content to lowercases and remove punctuation.
    """

    corpus = []
    
    for item in docs:
        doc_file=open(item,'r')
        while 1:
            line=doc_file.readline().replace('.','').replace(',','').strip().lower()
            if len(line)==0:
                break
            corpus.append(line)
        doc_file.close()
    
    """
    Loop `corpus`. Append the terms in each doc into the `bag_of_words` array. The terms in `bag_of_words` 
    should be unique which means before adding each term you need to check if it's already added to the array.
    In addition, check if each term is in the `stop_words` array. Only append the term to `bag_of_words`
    if it is not a stop word.
    """

    bag_of_words = []
    
    for item in corpus:
        for element in item.split():
            if element not in bag_of_words and element!='' and element not in stop_words:
                bag_of_words.append(element)
    
    """
    Loop `corpus` again. For each doc string, count the number of occurrences of each term in `bag_of_words`. 
    Create an array for each doc's term frequency and append it to `term_freq`.
    """
    
    term_freq = []
    
    for item in corpus:
        freq_2=[]
        for element in bag_of_words:
            if element in item:
                freq_2.append(1)
            else:
                freq_2.append(0)
        term_freq.append(freq_2)
    
    # Now return your output as an object
    return {
        "bag_of_words": bag_of_words,
        "term_freq": term_freq
    }
    
```

```python
import re

def remove_tags(text):
    """
    Removes every tag existing in an html document
    """
    TAG_RE=re.compile(r'<[^>]+>')
    return TAG_RE.sub('', text)

def remove_java(text):
    """
    If a line contents javascript it will substitue every character by a gap
    """
    if text.strip().startswith('= javascript_include_tag'):
        return ''
    else:
        return text

def remove_keys(text):
    """
    Removes {} elements
    """
    key_RE=re.compile('{.*}')
    return key_RE.sub('',text)

def remove_description(text):
    if text.strip().startswith('"'):
        return ''
    else:
        return text
    
def remove_more_elemts(text):
    """
    Multiple statements to clean a huge variety of characters
    """
    nn0=re.sub('&#39;','´',text) # For contractions like you´ll
    nn1=re.sub('[^a-zA-Z0-9$€´ \n\.]', ' ', nn0)
    return nn1

def remove_undessired_html_elements(text):
    """
    Cleans everything inside an html doc
    As input it takes a string to evaluate it
    """
    text1=remove_tags(text)
    text2=remove_java(text1)
    text3=remove_keys(text2)
    text4=remove_description(text3)
    text5=remove_more_elemts(text4)
    return text5.strip()

wp=open('www.coursereport.com_ironhack.html','r')
counter=0
while 1:
    counter+=1
    line=wp.readline()
    linek=remove_undessired_html_elements(line)
    if len(line)==0:
        break
    print(counter," :  ",linek)
wp.close()
```
